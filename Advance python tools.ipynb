```python
import numpy as np
```

```python
#1. Create a 1D NumPy array with integers from 1 to 20. Perform the following operations: 
arr1= np.arange(1,21)
print("Array:",arr1)
```

```python
#a. Calculate the sum, mean, median, and standard deviation of the elements in the array. 
print("sum:",np.sum(arr1))
print("mean:",np.mean(arr1))
print("median:",np.median(arr1))
print("standard_deviation:",np.std(arr1))
```

```python
#b. Find the indices of elements greater than 10 in the array. 
idc=np.where(arr1>10)[0]
print("greater than 10:",idc)
```

```python
#2. Create a 2D NumPy array of shape 4 X 4 with numbers ranging from 1 to 16. 
#a. Print the array.
arr2=np.arange(1,17).reshape(4,4)
print('2D Array:\n',arr2)
```

```python
# b. Find the transpose of the array. 
print("transpose:",arr2.T)
```

```python
# c. Row-wise and column-wise sums
print("Row-wise sum:",    np.sum(arr2, axis=1))
print("Column-wise sum:", np.sum(arr2, axis=0))
```

```python
#3. Create two 3 X 3 arrays filled with random integers between 1 and 20. 
# Two 3x3 random integer arrays (1 to 20)
np.random.seed(42)  # for reproducibility
a = np.random.randint(1, 21, size=(3, 3))
b = np.random.randint(1, 21, size=(3, 3))
print("Array A:\n", a)
print("Array B:\n", b)

# a. Element-wise operations
print("Addition:\n",       a + b)
print("Subtraction:\n",    a - b)
print("Multiplication:\n", a * b)

# b. Dot product
print("Dot Product:\n", np.dot(a, b))
```

```python
#4. Reshape 1D (size 12) → 3×4, then slice first 2 rows and last 2 columns
# 1D array of 12 elements, reshaped to 3x4
arr = np.arange(1, 13).reshape(3, 4)
print("Reshaped Array:\n", arr)

# Slice: first 2 rows (:2) and last 2 columns (-2:)
sliced = arr[:2, -2:] 
print("Sliced (first 2 rows, last 2 cols):\n", sliced)
```

1. Create a DataFrame with the following data:  
  


  data = {
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
  }
  
  a. Print the first five rows of the DataFrame.  
  b. Get the summary statistics of the 'Age' and 'Salary' columns.  
  c. Calculate the average salary of employees in the 'HR' department.  


2. Add a new column, 'Bonus', which is 10% of the salary.

```python
import pandas as pd
```

```python
# 1. Create a DataFrame with the following data:
data = { 
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'], 
    'Age': [24, 27, 22, 32, 29], 
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'], 
    'Salary': [45000, 54000, 50000, 62000, 47000] 
}
```

```python
df=pd.DataFrame(data)
```

```python
#a. Print the first five rows of the DataFrame.
print(df.head())
```

```python
# b. Get the summary statistics of the 'Age' and 'Salary' columns.
print(df[['Age','Salary']].describe())
```

```python
# c. Average salary in HR department
hr_avg=df[df['Department']=='HR']['Salary'].mean()
print("avg salary\n",hr_avg)
```

```python
#2.Add a new column, 'Bonus', which is 10% of the salary.
df['Bonus'] = df['Salary'] * 0.10
print(df)
```

```python
#3. Filter the DataFrame to show employees aged between 25 and 30.
filtered=df[(df['Age']>25)&(df['Age']<30)]
print("Age between 25 and 30:\n",filtered)
```

```python
#4. Group the data by 'Department' and calculate the average salary for each department.
dept_avg_sal= df.groupby('Department')['Salary'].mean().reset_index()
print("Averagr salary department wise:\n",dept_avg_sal)
```

```python
#5. Sort the DataFrame by 'Salary' in ascending order and save the result to a new CSV file.
sorted_salary=df.sort_values('Salary',ascending=True)
print(sorted_salary)
```

```python
sorted_salary.to_csv('sorted_salary.csv',index=False)
print("updated")
```

```python
#a. Plot the data.
#b. Customize the plot by adding a title, axis labels, and a grid.
x = [1, 2, 3, 4, 5] 
y = [10, 15, 25, 30, 50]

plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o', color='blue', linewidth=2)
plt.title('Simple Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()
```

```python
#3. Create a pie chart to represent the percentage distribution of a company’s revenue from different regions:
#a. Create a pie chart with the region names as labels.
#b. Highlight the region with the highest revenue.
regions = ['North America', 'Europe', 'Asia', 'Others'] revenue = [45, 25, 20, 10]regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# explode highlights the first slice (North America)
explode = (0.1, 0, 0, 0)

plt.figure(figsize=(6, 6))
plt.pie(revenue,
        labels=regions,
        explode=explode,
        autopct='%1.1f%%',
        startangle=140)
plt.title('Revenue Distribution by Region')
plt.show()
```

```python
4. #Generate a histogram to show the frequency distribution of randomly generated integers between 1 and 100 (sample size = 1000).
np.random.seed(0)
data = np.random.randint(1, 101, size=1000)

plt.figure(figsize=(8, 4))
plt.hist(data, bins=20, color='steelblue', edgecolor='black')
plt.title('Frequency Distribution (1–100, n=1000)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
